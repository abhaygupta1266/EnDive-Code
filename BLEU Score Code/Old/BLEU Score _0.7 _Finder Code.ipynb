```python
import pandas as pd
import os

# Function to calculate percentage of BLEU scores < 0.7
def calculate_percentage_below_threshold(nonfiltered_file_path, filtered_file_path, output_file_path):
    # Load both datasets
    nonfiltered_df = pd.read_csv(nonfiltered_file_path)
    filtered_df = pd.read_csv(filtered_file_path)

    # Calculate percentages
    total_count = len(nonfiltered_df)
    below_threshold_count = len(filtered_df)

    percentage_below_threshold = (below_threshold_count / total_count) * 100

    # Write percentage to a text file
    with open(output_file_path, 'w') as f:
        f.write(f"Total Samples: {total_count}\n")
        f.write(f"Samples Below Threshold (<0.7): {below_threshold_count}\n")
        f.write(f"Percentage Below Threshold: {percentage_below_threshold:.2f}%\n")

    print(f"Percentage of BLEU scores < 0.7 saved to: {output_file_path}")

# Results for all dialects and tasks
dialects = ["JamE", "EAAVE", "IndE", "CollSgE", "ChcE"]
results = []

for dialect in dialects:
    results.extend([
        # Logic Bench MCQ
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench MCQ/logic_bench_mcq_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench MCQ/logic_bench_mcq_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench MCQ/logic_bench_mcq_percentage.txt"
        },
        # Logic Bench YN
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench YN/logic_bench_yn_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench YN/logic_bench_yn_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/Logic Bench YN/logic_bench_yn_percentage.txt"
        },
        # SVAMP
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/SVAMP/svamp_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/SVAMP/svamp_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/SVAMP/svamp_percentage.txt"
        },
        # GSM8K
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/GSM8K/gsm8k_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/GSM8K/gsm8k_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/GSM8K/gsm8k_percentage.txt"
        },
        # MBPP
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/MBPP/mbpp_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/MBPP/mbpp_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/MBPP/mbpp_percentage.txt"
        },
        # HumanEVAL
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/HumanEVAL/humaneval_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/HumanEVAL/humaneval_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/HumanEVAL/humaneval_percentage.txt"
        },
        # FOLIO
        {
            "nonfiltered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/FOLIO/folio_bleu_scores.csv",
            "filtered_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/FOLIO/folio_filtered_bleu_scores.csv",
            "output_file_path": f"/content/drive/MyDrive/!!Multi-AAVENUE/BLEU Score Datasets/{dialect}/FOLIO/folio_percentage.txt"
        }
    ])

# Process each result
for result in results:
    calculate_percentage_below_threshold(
        result["nonfiltered_file_path"],
        result["filtered_file_path"],
        result["output_file_path"]
    )

print("Processing completed.")
```
